3.	Write a python program to demonstrate translation of the image from present position.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread(r"/content/flowerip.jpg")
height,width=img.shape[:2]
q_h,q_w=height/4,width/4
T=np.float32([[1,0,q_w],[0,1,q_h]])
img_trans=cv2.warpAffine(img,T,(height,width))
plt.imshow(cv2.cvtColor(img_trans,cv2.COLOR_BGR2RGB))
plt.show()
```

4.	Write a python program to Enhance an image using Image Arithmetic and Logical Operations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1=cv2.imread(r"/content/flowerip.jpg")
img2=cv2.imread(r"/content/rgbimg.png")
img1=cv2.resize(img1,(200,250))
img2=cv2.resize(img2,(200,250))
addImage = cv2.addWeighted(img1,.6,img2,.3,.2)
plt.imshow(cv2.cvtColor(addImage,cv2.COLOR_BGR2RGB))
plt.show()
addImage = cv2.subtract(img1,img2)
plt.imshow(cv2.cvtColor(addImage,cv2.COLOR_BGR2RGB))
plt.show()
addImage = cv2.bitwise_xor(img1,img2,mask=None)
plt.imshow(cv2.cvtColor(addImage,cv2.COLOR_BGR2RGB))
plt.show()
```

b.	Downscale the image by resizing with custom height and width while not preserving the aspect ratio.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
path=r"/content/flowerip.jpg"
image=cv2.imread(path)
#down scaling
down_width=400
down_height=200
down_point=(down_width,down_height)
resized_down=cv2.resize(image,down_point,interpolation=cv2.INTER_LINEAR)
cv2_imshow(resized_down)
#upscaling
up_width=200
up_height=400
up_point=(up_width,up_height)
resized_down=cv2.resize(image,up_point,interpolation=cv2.INTER_LINEAR)
cv2_imshow(resized_down)
```

c.	Scale the image vertically and horizontally and display all three images.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
path=r"/content/flowerip.jpg"
image=cv2.imread(path)
#horizontal
width=300
dim_size=(width,image.shape[0])
hor_img=cv2.resize(image,dim_size,interpolation=cv2.INTER_AREA)
cv2_imshow(hor_img)
#vertical
height=600
dim2_size=(height,image.shape[1])
ver_img=cv2.resize(image,dim2_size,interpolation=cv2.INTER_AREA)
cv2_imshow(ver_img)
```

d.	Downsampling and Upsampling an image while preserving the aspect ratio.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
path=r"/content/flowerip.jpg"
image=cv2.imread(path)
img_half=cv2.resize(image,(0,0),fx=0.5,fy=0.5)
cv2_imshow(img_half)
img_up=cv2.resize(image,(0,0),fx=2,fy=2)
cv2_imshow(img_up)
```
